# Python program to calculate the difference between two dates

date1 = pd.to_datetime(date_str1)
date2 = pd.to_datetime(date_str2)

time_difference = date2 - date1

print(f"Difference in days: {time_difference.days}")
print(f"Difference in hours: {time_difference.seconds // 3600}")
print(f"Difference in minutes: {time_difference.seconds % 3600 // 60}")

# Q9
# Python program to read a CSV file containing categorical data
file_path = input("Enter the CSV file path: ")
column_name = input("Enter the column name to convert to categorical data type: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

df_categorical = pd.read_csv(file_path)
df_categorical[column_name] = pd.Categorical(df_categorical[column_name], categories=category_order, ordered=True)
sorted_data = df_categorical.sort_values(by=column_name)
print(sorted_data)
